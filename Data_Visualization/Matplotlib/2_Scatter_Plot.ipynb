```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
import pandas as pd

plt.style.use('seaborn-whitegrid')
```

Scatter plot can be done with 
- plt.plot()
- plot.scatter()

## Scatter plots with `plt.plot()`

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, c='r')      # this produce a line plot with red line
plt.plot(x, y, 'o', c='k') # this produce a scatter plot with black dots
```

Or with only one argument `'-og'`, we can connect the line and dots:
- `'-'` defines line style
- `'o'` defines dot style
- `'g'` defines color as green

```python
plt.plot(x, y, '-og')
```

### Point Styles

```python
rng = np.random.RandomState(0)
marker_list = ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']

plt.figure(figsize=(8,8))
for marker in marker_list:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1, fontsize=16)
plt.xlim(0, 1.8)
```

### Use arguments inside of `plt.plot()`

```python
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2)
```

## Scatter Plot with `plt.scatter()`

```python
plt.scatter(x, y, marker='o')
```

The difference between plt.scatter() and plt.plot() is that it can be used to create scatter plots where the properties of each individual point(size, face color, edge color, etc) can be individually controlled or mapped towards data, thus provide more dimensions.

Let's plot the iris dataset:

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'])
```

Change the styles based on data point's features:

```python
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'],
            alpha = 0.3,
            s = 100*df['petal width (cm)'], # the size of the dot reflects pedal width
            c = iris.target,
            cmap='viridis')
plt.xlabel(df.columns[0])
plt.ylabel(df.columns[1])
plt.legend()
```
