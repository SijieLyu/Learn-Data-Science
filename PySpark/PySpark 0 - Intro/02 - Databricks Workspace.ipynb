#Lesson 02 - Databricks Workspace

## Getting Started with Databricks.

Databricks is a cloud-based data analytics platform created by the developers of Spark. We will use this service in this class to run Spark in an installation-free environment. The code we write in Databricks will be ran on clusters of virtual machines. The virtual machines that make-up our clusters will be [AWS EC2 instances](https://docs.aws.amazon.com/AWSEC2/latest/UserGuide/concepts.html). We will discuss clusters more in the next lesson.

You should have received an invitation to join a Databricks deployment that has been created specifically for this course. When you click the link provided in that email, you will be taken to a page that asks you to enter your name and set a password for your account. Please enter your real first and last name. 

After your account is created, you will be taken to the Databricks dashboard page.

## Workspace Setup

A **notebook** in Databricks is a document consisting of cells containing either executable code or formatted text, similar to a Jupyter Notebook. The lessons in this course will be delivered in a notebook format. Additionally, any homework assignment or project you work on with involve creating a notebook. Your workspace is where you will store all of the notebooks you create for this course. To view the contents of your workspace, following the steps provided below.

1. Click the **Workspace** icon in the left sidebar. This will pop out the workspace browser.
2. If you wish to keep the workspace browser open, you can click the pin icon in the upper right corner of the browser.
3. Click on **Users** to expand the contents of that directory.
4. Then click on your email address (this should have a home icon next to it).  This shows you the contents of your user directory.

Initially, your user directory will contain only a single item named **Trash**. We will walk through a few initial steps to set up and organize your workspace.

### Creating Folders

We will create a few folders to help keep your workspace organized. To create the first folder, right click in the panel associated with your user account, just below the **Trash** item. Then click **Create**, and then **Folder**. In the pop-up that appears, enter the name **Homework** for your first folder. Then repeat these steps to create a folder named **Projects** and a folder named **Sandbox**. 

The **Homework** folder will contain your homework submissions, while the **Projects** folder will be used to store your projects. The **Sandbox** folder will be for your own personal use so that you can practice and explore on your won.

### Copying Lessons Folder

The lesson content for this course has been provided to you in the form of several notebooks. These are stored in a folder named **Lessons** in the top level of the workspace browser. To find these, you might need to click the left arrow (**`<`**) on the left side of the browser. You are able to view the content of these notebooks, but are not able to edit them or execute the code cells they contain. However, you are able to create your own copy that you can use as you would like. I recommend that you do this since the best way to learn in this course is by actually running and experimenting with the code inside the lesson notebooks.

To copy the **Lessons** folder, click the down arrow to the of the name of the folder, then click **Clone**. In the window that appears, set the name of the new folder to **Lessons**. Confirm that Users is selected in the leftmost panel and your email address is selected in the middle panel. Then click **Clone**.

When you navigate back to your user directory, you should see five items: **Trash**, **Homework**, **Lessons**, **Projects**, and **Sandbox**.
