#Lesson 03 - Databricks Clusters

## Creating a Cluster

The first task you will need to complete within Databricks is the create a cluster. A **cluster** is a network of virtual machines that will be used to execute our Spark jobs. To create a cluster, please following the steps provided below.

1. From the Databricks dashboard page, click **New Cluster** under **Common Tasks**.
2. Provide your cluster with a name. I recommend using a name in the following format for your first cluster: **`username_r5_large`**, replacing "`username`" with your Maryville username. The string `r5_large` indicates the type of AWS EC2 virtual machines that will constitute your first cluster. Using this naming format will help you to keep track of your clusters as you create more in the future. Leave all other settings with their default values set. 
3. Click **Create Cluster**.
4. This will take you to a page with information about the cluster. For now, click on the **Clusters** link in the upper left corner. This will take you back to the clusters page. You should see a cluster with the name you provided. Initially, the state of the cluster will show as **Pending**. You may need to wait a few minutes before  the cluster is in a usable state. Eventually, the cluster state should change to **Running**. You might need to refresh your browser to see the updated state. When you see the state listed as **Running**, the cluster is ready to use.

## Starting and Stopping a Cluster

Your clusters will use AWS credit whenever they are running. To preserve this credit, you should stop your cluster whenever you are done using it. Clusters will also automatically stop after 30 minutes of inactivity, but it is still good practice to manually stop them when they are no longer in use. Before we move on to discussing how to use a cluster, lets practice stopping and starting a cluster.

### Stopping A Cluster

To stop a cluster, click on the **Clusters** icon in the left sidebar. Then click on the name of the running cluster. This takes you to the management page for that cluster. Click the **Terminate** button at the top of the page. Do not click **Delete** unless you want to permanently delete this particular cluster. You will be presented with a confirmation pop-up. Click **Confirm**. Click on the **Clusters** sidebar icon again. You should see the State of the cluster listed as **Terminating**. Eventually, this will change to **Terminated**.

### Starting A Cluster

Whenever you return to work within Databricks, you will need to restart your cluster. To start a cluster, click on the **Clusters** icon in the left sidebar. Then click on the name of the cluster you wish to start. On the cluster management page, click the **Start** button, and then click **Confirm** in the pop-up that appears. Click the **Clusters** icon to return to the clusters page, and you will see the state of the cluster listed as **Pending**. Eventually the state will change to **Running**, at which point the cluster will again be ready to use.
