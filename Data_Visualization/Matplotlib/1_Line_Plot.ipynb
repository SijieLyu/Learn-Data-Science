```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-whitegrid')
```

## Two Ways to Plot
- ax, fig
- plt

Start by creating figure and axes. In the simplest form:

```python
fig = plt.figure()
ax = plt.axes()
```

Create a simple sinusoid:

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(1, 10, 1000)
ax.plot(x, np.sin(x))
```

Or we can use pylab interface:

```python
plt.plot(x, np.sin(x))
```

add cosine:

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

## Adjust Line Colors and Styles

```python
plt.plot(x, np.sin(x - 0), color='blue', label = 'sine(x-0)')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g', label = 'sine(x-1)')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75', label = 'sine(x-2)')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44', label = 'sine(x-3)')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3), label = 'sine(x-4)') # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse', label = 'sine(x-5)')  # all HTML color names supported
plt.legend()
```

## Adjust Line Style
- argument "linestyle = 'whatever line style of your choice' "

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')
```

For short, we can use the following codes:

```python
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

## Linestyle + Color
- linestyle and color can be combined into a single non-keyword argument to the `plt.plot()` function:

```python
plt.plot(x, x + 0, '-g', label='f(x)=x')   # solid green
plt.plot(x, x + 1, '--c', label= 'f(x)=x+1')  # dashed cyan
plt.plot(x, x + 2, '-.k', label='f(x)=x+2')  # dashdot black
plt.plot(x, x + 3, ':r', label = 'f(x)=x+3')  # dotted red
plt.legend()
```

## Adjust Axes Limits
- plt.xlim()
- plt.ylim()

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

`plt.axis()`: this method allows set the x and y limits with a single call, by passing specifies `[xmin, xmax, ymin, ymax]`:

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
```

`plt.axis()`: can also tight the bounds around current plot, or ensure equal ratio by passing aruguments like
- 'tight'
- 'equal'

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

## Labeling Plots
- `plt.title("plot title")`: adds a title of the plot
- `plt.xlabel("x-axis label")`: adds x-axis label
- `plt.ylabel("y-axis label)`: adds y-axis label
- or we can pass the label using `label` argument in `plt.plot()`

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)') #sin(x), with solid green line, labeled
plt.plot(x, np.cos(x), ':b', label='cos(x)') #cos(x), with dotted blue line, labeled
plt.axis('equal')

plt.legend() # add legend
```

## Convert between Object-oriented Method Style and MATLAB-Style
- `plt.xlabel()` -> `ax.set_xlabel()`
- `plt.ylabel()` -> `ax.set_ylabel()`
- `plt.xlim()` -> `ax.set_xlim()`
- `plt.ylim()` -> `ax.set_ylim()`
- `plt.title()` -> `ax.set_title()`

Or using oop interface with `ax.set()` to pass multiple arguments about limits, styles and titles:

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot')
```
